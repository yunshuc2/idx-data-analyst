```python
import pandas as pd
import glob
```

```python
#SOLD
file_path = "/Users/yunshucao/Desktop/yunshu/IDX Intern/CRMLSSold202602.csv"

df = pd.read_csv(file_path)
print(df.columns)
```

```python
print("original length:", len(df))

df = df.drop_duplicates()
df["CloseDate"] = pd.to_datetime(df["CloseDate"], errors="coerce")
df = df.dropna(subset=["CloseDate", "ClosePrice", "LivingArea"])
df = df[df["LivingArea"] > 0]


df["year"] = df["CloseDate"].dt.year
df["month"] = df["CloseDate"].dt.month
df["year_month"] = df["CloseDate"].dt.to_period("M").astype(str)

df["price_per_sqft"] = df["ClosePrice"] / df["LivingArea"]


df = df.rename(columns={
    "CloseDate": "sold_date",
    "ClosePrice": "sold_price",
    "LivingArea": "living_area"
})


output_path = "/Users/yunshucao/Desktop/yunshu/IDX Intern/sold_clean_202602.csv"
df.to_csv(output_path, index=False)

print("output path:", output_path)
print(df[["sold_date", "sold_price", "living_area", "price_per_sqft"]].head())
```

```python
#LISTED
file_path = "/Users/yunshucao/Desktop/yunshu/IDX Intern/CRMLSListing202602.csv"

df = pd.read_csv(file_path)
print(df.columns)
```

```python

print("original length(Listing):", len(df))
df = df.drop_duplicates()
df["ListingContractDate"] = pd.to_datetime(df["ListingContractDate"], errors="coerce")
df = df.dropna(subset=["ListingContractDate", "ListPrice", "LivingArea"])
df = df[df["LivingArea"] > 0]
df["year"] = df["ListingContractDate"].dt.year
df["month"] = df["ListingContractDate"].dt.month
df["year_month"] = df["ListingContractDate"].dt.to_period("M").astype(str)

df["price_per_sqft"] = df["ListPrice"] / df["LivingArea"]

df = df.rename(columns={
    "ListingContractDate": "listing_date",
    "ListPrice": "list_price",
    "LivingArea": "living_area",
    "DaysOnMarket": "days_on_market",
    "BathroomsTotalInteger": "bathrooms",
    "BedroomsTotal": "bedrooms"
})

keep_cols = [
    "ListingKey",
    "ListingId",
    "listing_date",
    "year",
    "month",
    "year_month",
    "list_price",
    "living_area",
    "price_per_sqft",
    "days_on_market",
    "City",
    "PostalCode",
    "CountyOrParish",
    "PropertyType",
    "PropertySubType",
    "bedrooms",
    "bathrooms",
    "YearBuilt",
    "LotSizeSquareFeet",
    "GarageSpaces",
    "Stories",
    "MlsStatus",
    "ListOfficeName",
    "ListAgentFullName",
    "Latitude",
    "Longitude",
    "UnparsedAddress"
]

keep_cols = [col for col in keep_cols if col in df.columns]

df = df[keep_cols]

output_path = "/Users/yunshucao/Desktop/yunshu/IDX Intern/listed_clean_202602.csv"
df.to_csv(output_path, index=False)

print("output path:", output_path)
print(df.head())
```
